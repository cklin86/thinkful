##  Renewable Energy Production

The the rise of emerging market economies like China and India have increased the need for sustainable clean energy sources. The ongoing threat of global warming and rising temperatures have also driven developed nations to invest further in clean energy. It is beleived that renewable energy sources will provide 85% of global electricity production in 2050, led by solar PV and onshore wind.


```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
# the following instruction is just for Jupyter; it tells Jupyter to display graphics generated by matplotlib
%matplotlib inline
```

```python
df=pd.read_csv("UNdata_Export_Electricity - total solar production.csv",sep=",")
```

```python
df.columns = ["Country","Commodity","Year","Unit","Quantity","Quantity Footnotes"]
```

```python
df.groupby(['Year']).groups.keys()
```

```python
# Get the sum of Global Total Solar Production - Kilowatt-hours,(M)
world_solar = df.groupby('Year')['Quantity'].sum() 
print(world_solar)
```

Worldwide growth of photovoltaics has been an exponential curve between 1990â€“2014. During this period of time, photovoltaics (PV), also known as solar PV, evolved from a niche market of small scale applications to a mainstream electricity source. When solar PV systems were first recognized as a promising renewable energy technology, programs, such as feed-in tariffs, were implemented by a number of governments in order to provide economic incentives for investments. As a consequence, cost of solar declined significantly due to Experience curve effects like improvements in technology and economies of scale.

The chart below shows the the exponential surge in global solar production from 1990 to 2014.

```python
my_plot = world_solar.plot(kind='bar',legend=None,title="Global Solar Production - Killowatt-hours (m)")
my_plot.set_xlabel("Year")
my_plot.set_ylabel("Killowatt-hours (m)")
```

```python
USA = df.loc[df['Country']=='United States',['Year','Quantity']]
USA = USA.sort_values(['Year','Quantity'],ascending=[True,False])
print(USA)
```

This worldwide surge in solar production has also been mirrored in the United States, with domestic solar production rising rapidly during this same period.

It is often believed that a decrease in economic activity would lead to governments and industries to turn back to cheaper non-renewable sources of energy. However the Great Recession of 2007-2009, it is evident that this had little impact on global solar production. A fall in global GDP in 2009 and a slow recovery in the following years, seemed to do little to affect the rapid growth in solar production worldwide. This was partially attributed to stimulus from governements around the world including the U.S. In the United States under Obama administration's stimulus packages, a 30 percent cash rebate for individuals or companies installing solar energy systems, and provided loan guarantees that will brought down interest rates for companies like BrightSource, who were building new solar installations.



```python
all_energy=pd.read_csv("All_energy.csv",sep=",")
```

```python
all_energy.columns = ["Year","Coal","Petroleum","Natural Gas","Other Gas","Nuclear","Hydro Conventional","Hydro Pumped","Geothermal","Wind","WTI"]
all_energy = all_energy.sort_values(['Year'],ascending=[True])
all_energy.info
```

Lower fossil fuel prices have also often been attributed to lower production and switching incentives to renewable energy. It is believed that for example lower price of crude oil, will lessen the incetive to swithch to renewable sources like solar and wind. Historical data over the last 10 years however seems to prove otherwise. 

```python
plot_wind=all_energy.plot(x='Year',y='Wind',kind='bar',legend=None,color='grey',title= "United States - Wind ('000s Megawatthours)")
plot_wind.set_xlabel("Year")
plot_wind.set_ylabel("000's Megawatthour")

plot_wti=all_energy.plot(x='Year',y='WTI',kind='bar',legend=None,color='darkgreen',title= "Crude Oil - WTI - USD per Barrel")
plot_wti.set_xlabel("Year")
plot_wti.set_ylabel("USD per Barrel")
```

The recent sharp fall in crude oil prices in from over \$100 to \$48 in 2014 to 2015, still saw a steady increase wind power production in the United States. In fact, over the last 10 years, the price of crude oil has been volatile ranging from  a peak of almost \$100 to less than \$50 over this period. Wind power output however has been increasing year on year. 

As far as solar and wind go, the impact from lower oil prices is zero in North America and Europe, where power prices do not have any link to oil. In some emerging markets for wind and solar, however, such as the Middle East and North Africa, and even in a few established markets such as Japan, diesel does still play a bigger role in power supply â€“ but not big enough to matter much considering current renewable energy prices.
